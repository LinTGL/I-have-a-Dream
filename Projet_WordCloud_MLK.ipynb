```python
import string

#Utilisation de la bibliothèque 'string' pour supprimer ponctuations et stopwords

#transformations des tokens en miniscules pour faciliter le traitements

tokens = nltk.word_tokenize(MLK_txt.lower()) 
#on l'enregistre dans une variable afin de la réutiliser pour gérer la ponctuatio et les stopwords

token_clean = []
for items in tokens:
    if items not in nltk.corpus.stopwords.words("french") and items not in string.punctuation:
        token_clean.append(items)

token_clean
```

LEMMATIZER

```python
import spacy

nlp = spacy.load('fr_core_news_sm') #en anglais : ('en_core_web_sm')
```

```python
#transforme ma liste de tokens en chaine de caractère pour le traitement avec lemmatizer

token_ok = " ".join(token_clean)
token_ok
```

```python
#application du lemmatizer

lemma_tok = nlp(token_ok)
for word in lemma_tok:
    print(word.lemma_)
```
